```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


drug0_top1 = pd.read_csv("temp_data/finetune0_top1score.csv")
drug0_mean = pd.read_csv("temp_data/finetune0_meanscore.csv")
drug1_top1 = pd.read_csv("temp_data/finetune1_top1score.csv")
drug1_mean = pd.read_csv("temp_data/finetune1_meanscore.csv")
drug2_top1 = pd.read_csv("temp_data/finetune2_top1score.csv")
drug2_mean = pd.read_csv("temp_data/finetune2_meanscore.csv")
```

```python
fig = plt.figure(figsize = (12,4))

ax1 = fig.add_subplot(131)
line01, = plt.plot("Step", "Value", data=drug0_top1, color="blue")
line02, = plt.plot("Step", "Value", data=drug0_mean, color="green", ls='--', linewidth=2)
ax1.set_xlabel("RL Fine-tuning Steps\n\n(a) Celecoxib", fontsize=15, fontproperties="Times New Roman")
ax1.set_ylabel("Scores", fontsize=15, fontproperties="Times New Roman")
ax1.set_ylim(-0.05, 1.05)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.legend([line01, line02], ["top1 score", "mean score"], prop={"family": "Times New Roman", "size": 15})
ax1.grid(ls=':')

ax2 = fig.add_subplot(132)
line11, = plt.plot("Step", "Value", data=drug1_top1, color="blue")
line12, = plt.plot("Step", "Value", data=drug1_mean, color="green", ls='--', linewidth=2)
ax2.set_xlabel("RL Fine-tuning Steps\n\n(b) Troglitazone", fontsize=15, fontproperties="Times New Roman")
# ax2.set_ylabel("Scores", fontsize=15, fontproperties="Times New Roman")
ax2.set_ylim(-0.05, 1.05)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.legend([line11, line12], ["top1 score", "mean score"], prop={"family": "Times New Roman", "size": 15})
ax2.grid(ls=':')

ax3 = fig.add_subplot(133)
line21, = plt.plot("Step", "Value", data=drug2_top1, color="blue")
line22, = plt.plot("Step", "Value", data=drug2_mean, color="green", ls='--', linewidth=2)
ax3.set_xlabel("RL Fine-tuning Steps\n\n(c) Thiothixene", fontsize=15, fontproperties="Times New Roman")
# ax3.set_ylabel("Scores", fontsize=15, fontproperties="Times New Roman")
ax3.set_ylim(-0.05, 1.05)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.legend([line21, line22], ["top1 score", "mean score"], loc='lower right', prop={"family": "Times New Roman", "size": 15})
ax3.grid(ls=':')

plt.tight_layout()
plt.savefig("finetune_scores.pdf")
```
